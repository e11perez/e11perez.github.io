# Abstract

In our project, we investigate whether there is a relationship between league affiliation (American League vs. American League) and the batting performance of MLB outfielders from the years 2015 to 2024. We focus on three key variables, batting average(Avg), on-base percentage(OBP), and on-base plus slugging(OPS) to clean and analyze player data statistics to better understand our hypothesis. In order to clean the and limit the amount of data needed, we filtered the sample data to contain players who played at least 81 games per season to ensure play consistency. Our exploratory data analysis included scatter plots, linear regression models, and t-tests to discover patterns between the league, player characteristics, and batting performance. In the end, our results showed that league affiliations are not a significant factor when predicting batting success, differences between leaguers across most seasons remained without a statistically significant difference. However, we found that the strongest predictor was the number of batting hits compared to age and game players which shows no impact. Due to the events of 2020 and the environmental factors related to COVID-19, we decided to treat this year as an outlier, given the insufficient data as a result of the short season. Additionally, we found that during the year 2022, the National League introduced a new rule, the Designated Hitter (DH) rule, which likely contributed to the change in performance across leagues. Overall, this study suggests that players' ability rather than league affiliation is the main driving force behind all MLB outfielders within the last decade. 

# Research Question

From the 2015-2024 MLB seasons, is there a correlation between outfielders' batting performance (Avg, OPS, OBP) and the league that they play in?



## Background and Prior Work


America’s pastime, baseball, has been garnering years and years of history, big plays, exciting moments, and tough competition between players from all over the world. Major League Baseball was formally founded in 1903 with two leagues, the National League and the American League. The league has certainly increased over the years and reached its current 30-team size in 1998. Both have teams across the country and have tight competitions throughout the season, leading to the inevitable and exciting World Series. There are plenty of players who stay with one team with long contracts, and those who stay for short periods, depending on how their contracts are formed. There are 9 different field positions that a player could play: pitcher, catcher, first-baseman, second-baseman, shortstop, third-baseman, left-fielder, center-fielder, right-fielder. We are excluding the position of Designated Hitter from this project, seeing as their sole purpose is to hit, which may skew the results of our specific question. 

There are 3 specific metrics that we are interested in researching concerning our question -- batting average, OPS, and OBP. OPS stands for On-Base Plus Slugging, and OBP means On-Base Percentage. These are typically the main 3 stats that critics look at to determine the quality of a player, so we are also interested in researching them, but there are several other metrics available. 

Historically, there has been a notion that the National League (NL) was superior to the American League (AL), where spectators pointed to NL dominance in All-Star Games and overall better performance by NL players. A Society for American Baseball Research study comparing player performance across both leagues surprisingly found no significant difference in purported NL superiority1. Accounting for variables such as the designated hitter (DH) rule, league-wide average chances, and survivorship bias, the study illustrated that more players actually performed better in the AL than the NL, but marginally.
Previous research has indicated that a player’s defensive position can influence their offensive statistics. A study analyzing MLB data from the period of 1903 to 2018 showed a positive correlation between a player’s field position and their batting average, implying that certain positions may indeed have inherently better hitters2. Although there were no specific positions analyzed here, the study did highlight the importance of position in evaluating batting performance.

The introduction of the DH rule in the AL has proven to be impactful on league-wide batting statistics as well3. A study examining the effects of the DH rule found that, upon it being implemented, the AL experienced higher team batting averages compared to the NL. The individual batting averages of non-DH players remained constant, however. This implies that the DH can boost overall team offense, but does not necessarily enhance the batting performance of players in other positions.

We chose to focus specficially on outfielders to stay consistent with batting statistics. Outfielders usually have similar tasks when on the field, and are more focused on offense. Including all positions could potentially cause inconsistent batting statistics, as some positions prioritize batting more than others. Outfielders are somewhat unique in how consistent their batting is, and as such we chose to focus on their statistics.

We hope that through this research, we can clarify some misconceptions on which league performs better. Although many fans are impassioned by seeing their team succeed, how much of that apparent success actually holds when compared to other teams? Is there a definitive answer, at least through looking at outfielders, on which teams are *better*?

# Data Analysis (EDA)

We look at three main metrics of batting performance (OPS, OPB, Batting Average (BA)) in this project to determine if the league that the MLB outfielders are in affects their batting performance more than other statistics. In our project, these other statistics are the players' age, the number of games they played (G), the number of at bats (AB), and the number of hits they have (H). To find a relationship between each player statistic and their batting performances, we generated a series of scatter plots, where each individual player represents a single data point on the graph. We grouped each cloud by league, color coding players in the American League in red, and National League players in blue. The goal with this kind of data analysis is to see if the league that the outfielders play in has a stronger affect on their batting performances compared to variable like their age, number of games played, and so on. 

To better understand the underlying structure of our dataset, we visualized univariate (single-variable) distributions using histograms for key performance metrics such as OPS, batting average, OBP, and player characteristics like age and games played. These plots help us identify the shape of the data distribution — for example, whether the values are normally distributed, skewed, or contain outliers. Understanding whether a variable is normally distributed is important because many statistical models, including linear regression, assume that residuals (differences between actual and predicted values) are normally distributed. If a variable is heavily skewed or has extreme outliers, it might violate these assumptions and affect the validity of our model. For potential future steps, like standardization or hypothesis testing, knowing the distribution informs us whether transformations are needed.

From these resulting graphs, we can now transition from our EDA to more statistical inference. On each target variable, we run a multiple linear regression using `statsmodels`. From there, we extract the p-values from each graph to determine the relative effect of each variable on batting performance. We examine the p-values, and see that if the p-value < 0.05, this would indicate that the variable in question has a statistically significant relationship with the outcome. 

Below are the results for this analysis, going sequentially from the 2015 to the 2024 season. 

```python
import statsmodels.formula.api as smf
```

```python
# Combine AL and NL for each year into a single DataFrame
season_dfs = {
    2015: pd.concat([AL_df15, NL_df15], ignore_index=True),
    2016: pd.concat([AL_df16, NL_df16], ignore_index=True),
    2017: pd.concat([AL_df17, NL_df17], ignore_index=True),
    2018: pd.concat([AL_df18, NL_df18], ignore_index=True),
    2019: pd.concat([AL_df19, NL_df19], ignore_index=True),
    2020: pd.concat([AL_df20, NL_df20], ignore_index=True),
    2021: pd.concat([AL_df21, NL_df21], ignore_index=True),
    2022: pd.concat([AL_df22, NL_df22], ignore_index=True),
    2023: pd.concat([AL_df23, NL_df23], ignore_index=True),
    2024: pd.concat([AL_df24, NL_df24], ignore_index=True),
}

# List of performance metrics (dependent variables)
targets = ["BA", "OBP", "OPS"]

# List of numeric columns to ensure proper type
cols_to_numeric = ["Age", "G", "AB", "H", "BA", "OBP", "OPS"]

# Dictionary to store all model results
all_results = {}

# Loop through each year's dataset
for year, df in season_dfs.items():
    # Convert relevant columns to numeric, drop missing data
    df[cols_to_numeric] = df[cols_to_numeric].apply(pd.to_numeric, errors='coerce')
    df = df.dropna(subset=cols_to_numeric + ["League"])

    # Initialize yearly result container
    all_results[year] = {}

    # Run a regression for each performance metric
    for target in targets:
        formula = f"{target} ~ C(League) + Age + G + AB + H"
        model = smf.ols(formula=formula, data=df).fit()
        all_results[year][target] = model

        # Display regression summary
        print(f"\n=== {year} Regression for {target} ===")
        print(model.summary())
```

# Results

The results show that the league wasn’t a strong predictor or influence on the batting performance of outfielders. In most seasons, the league variable wasn’t statistically significant, and when it was (like in 2021 or 2022), the effect was very small.
